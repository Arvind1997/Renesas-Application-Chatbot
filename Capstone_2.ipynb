```python
import asyncio
asyncio.set_event_loop_policy(asyncio.DefaultEventLoopPolicy())
import nest_asyncio
nest_asyncio.apply()  # Allow nested event loops

from requests_html import AsyncHTMLSession
import asyncio

async def fetch_dynamic_content(url):
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender(timeout=20)  # Render JavaScript asynchronously
    return response.html.html
```

```python
url = "https://www.renesas.com/en/applications"

# Run the async function
html_content = asyncio.run(fetch_dynamic_content(url))
```

```python
soup = BeautifulSoup(html_content, 'html.parser')

categories = ['automotive', 'consumer-electronics', 'industrial', 'communications-infrastructure']

links = []
for i in categories:
  links.append(list(set([a["href"] for a in soup.find_all("a", href=True) if f'/en/applications/{i}' in a["href"]])))
links = sum(links, [])
```

```python
url = "https://www.renesas.com"

final_links = []

for i in links:
  html_content = asyncio.run(fetch_dynamic_content(url+i))
  soup = BeautifulSoup(html_content, 'html.parser')
  final_links.append(list(set([a["href"] for a in soup.find_all("a", href=True) if i in a['href']])))

final_links = list(set(sum(final_links, [])))
```

```python
async def extract_content_wrappers(final_links):

    schema = {

      "name": "Page Content 2",
      "baseSelector": "div.l-container",
      "fields": [
            {
              "name": "section_title",
              "selector": "div.application_header__title h1",
              "type": "text"
            },
            {
              "name": "Application Category",
              "selector": "div.pheader__content ol li:nth-child(2)",
              "type": "text"
            },
            {
              "name": "Application Sub category",
              "selector": "div.pheader__content ol li:nth-child(3)",
              "type": "text"
            },
            {
              "name": "Description",
              "selector": "div.description",
              "type": "nested",
              "fields": [
                {
              "name": "description",
              "selector": "p:nth-child(n)",
              "type": "text"
                },
                {
              "name": "System Benefits",
              "selector": "ul li",
              "type": "list",
              "fields": [
                {
              "name": "System Benefits",
              "type": "text"
                }
                ]
                }

                ]
            },
            {
              "name": "Applications",
              "selector": "div.field-applications",
              "type": "nested",
              "fields": [
                {
                  "name": "applications",
                  "selector": "ul li",
                  "type": "list",
                  "fields": [
                    {
                      "name": "applications",
                      "type": "text"
                    }
                ]
                }
                ]
            },
            {
              "name": "Winning Combinations Interactive Diagram",
              "selector": "div.field--name-field-media-image svg",
              "type": "html",
            }
        ]
      }

    extraction_strategy = JsonCssExtractionStrategy(schema, verbose=True)

    config = CrawlerRunConfig(
        cache_mode=CacheMode.BYPASS,
        extraction_strategy=extraction_strategy,
    )

    result = None
    async with AsyncWebCrawler(verbose=True) as crawler:
        result = await crawler.arun(
            url=final_links,
            config=config
        )

        if not result.success:
            print("Crawl failed:", result.error_message)
            return

        # Parse the extracted JSON
        data = json.loads(result.extracted_content)
        print(f"Extracted {len(data)} entries")
        print(json.dumps(data, indent=2) if data else "No data found")

    return result
```

```python
repo = []

for i in final_links[2:5]:
  res = asyncio.run(extract_content_wrappers(url+i))
  data = json.loads(res.extracted_content)
  repo.append(data)
```

```python
for i in range(0, len(repo)):
  svg_content = repo[i][0]['Winning Combinations Interactive Diagram']
  with open(f"/content/images/image_{i}.svg", "w", encoding="utf-8") as f:
    f.write(svg_content)
  # Convert SVG file to PNG
  cairosvg.svg2png(url=f"/content/images/image_{i}.svg", write_to=f"/content/images/image_{i}.png")

  # Convert SVG file to JPG (convert PNG first, then use PIL to save as JPG)
  img = Image.open(f"/content/images/image_{i}.png")
  img.convert("RGB").save(f"/content/images/image_{i}.jpg", "JPEG")

  repo[i][0]['Winning Combinations Interactive Diagram'] = f"/content/images/image_{i}.png"
```

```python
documents = []
for item in tqdm(range(0, len(repo))):
    text = json.dumps(repo[item], indent = 2)
    doc = Document(page_content=text, metadata={"source": "scraping"})
    documents.append(doc)
```

```python
embeddings = HuggingFaceEmbeddings(model_name = 'BAAI/bge-large-en', encode_kwargs = {'normalize_embeddings': True}, model_kwargs = {'device': 'cuda'})

vector_store = Chroma(persist_directory = './chromaDB', collection_name = 'job_description', embedding_function = embeddings)
```

```python
vector_store.add_documents(tqdm(documents))
```

```python
retriever = vector_store.as_retriever()
```

```python
llm = ChatOpenAI(api_key=OPENAI_API_KEY , model="gpt-3.5-turbo", temperature=0)
```

```python
### Contextualize question ###
contextualize_q_system_prompt = (
    "Given a chat history and the latest user question "
    "which might reference context in the chat history, "
    "formulate a standalone question which can be understood "
    "without the chat history. Do NOT answer the question, "
    "just reformulate it if needed and otherwise return it as is."
)
contextualize_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", contextualize_q_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)
history_aware_retriever = create_history_aware_retriever(
    llm, retriever, contextualize_q_prompt
)


### Answer question ###
system_prompt = (
    "You are an assistant for question-answering tasks. "
    "Use the following pieces of retrieved context to answer "
    "the question. If you don't know the answer, say that you "
    "don't know. "
    "Elaborate your answer based on the question and answer"
    "\n\n"

    "Context: {context}\n"
)
qa_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)
question_answer_chain = create_stuff_documents_chain(llm, qa_prompt)

rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)


### Statefully manage chat history ###
store = {}


def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]


conversational_rag_chain = RunnableWithMessageHistory(
    rag_chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history",
    output_messages_key="answer",
)


def get_answer(question):

  answer = conversational_rag_chain.invoke(
      {"input": question},
      config={
          "configurable": {"session_id": "ABc"}
      },  # constructs a key "abc123" in `store`.
  )["answer"]

  # Regex pattern to match the URL
  match = re.search(r"(/content/images/[\w\-.]+\.(jpg|png|svg|jpeg|gif))", answer)

  # Store the URL if found
  try:
    image_url = match.group(1) if match else None
    Image.open(image_url)
    response = re.sub(r'Winning Combinations .*', "diagram", answer)
    return response
  except:
    return answer
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
llm = ChatOpenAI(api_key = OPENAI_API_KEY, model="gpt-4o-mini")

history = []
image_url = ""

def llm_open(input):
  vectorizor = vector_store.as_retriever(search_kwargs={"k": 1})

  message = """
  Strictly Retrieve the Winning Combination Interactive Diagram url from the context and do not add or explain the concept.

  Context:
  {context}
  """

  prompt = ChatPromptTemplate.from_messages([("human", message)])

  rag_chain = {"context": vectorizor, "question": RunnablePassthrough()} | prompt | llm

  response = rag_chain.invoke(input)

  return response.content.strip()

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

def vision_model_response(text, file):
  image = Image.open(file)
  image.save(f"/content/images/image.png")
  client = OpenAI(api_key=OPENAI_API_KEY)
  base64_img = encode_image('/content/images/image.png')
  response = client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[
          {
              "role": "user",
              "content": [
                  {"type": "text", "text": "Explain the components in the image and provide detailed analysis on the design system"},
                  {
                      "type": "image_url",
                      "image_url": {
                          "url": f"data:image/png;base64,{base64_img}",
                      },
                  },
              ],
          }
      ],
      max_tokens=300,
  )
  os.remove("/content/images/image.png")
  return response.choices[0].message.content

def add_message(history, message):
    for img in message["files"]:
        history.append({"role": "user", "content": {"path": img}})
    if message["text"] is not None:
        history.append({"role": "user", "content": message["text"]})
    return history, gr.MultimodalTextbox(value=None, interactive=False), message['text']


def bot(input):
    history.append({"role": "user", "content": input})
    if(re.search(r"\bShow the design system\b|\bExplain the design system\b|\bdesign system\b|\bWinning Combination Interactive Diagram\b|\bdiagram\b", input)):
      image_url = llm_open(input)
      history.append({"role": "assistant", "content": {"path": image_url}})
    if(re.search(r"\bElaborate the diagram\b|\bExplain the components\b", input)):
      image_url = llm_open(input)
      response = vision_model_response(input, image_url)
      history.append({"role": "assistant", "content": response})
    response = get_answer(input)
    history.append({"role": "assistant", "content": response})
    time.sleep(1)
    return history

custom_css = """
#chatbot img {
    max-width: 800px;
    max-height: 500px;
    object-fit: contain;
}
"""

with gr.Blocks(css = custom_css, fill_height=True) as demo:
    message_state = gr.State()
    chatbot = gr.Chatbot(elem_id="chatbot", bubble_full_width=False, type="messages", scale = 1)

    chat_input = gr.MultimodalTextbox(
        interactive=True,
        file_count="multiple",
        placeholder="Enter message or upload file...",
        show_label=False,
        sources=["upload"],
    )

    chat_msg = chat_input.submit(
        add_message, [chatbot, chat_input], [chatbot, chat_input, message_state]
    )
    bot_msg = chat_msg.then(bot, inputs = [message_state], outputs = [chatbot], api_name="bot_response")
    bot_msg.then(lambda: gr.MultimodalTextbox(interactive=True), None, [chat_input])

if __name__ == "__main__":
    nest_asyncio.apply()
    demo.launch(debug = True)
```
